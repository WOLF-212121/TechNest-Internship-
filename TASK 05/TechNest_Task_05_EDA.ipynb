# **Exploratory Data Analysis**

# Distribution of Type in Dataset

```python
# Counts of Tvshows and movies
df1 = data.type.value_counts().rename('counts').reset_index()
aa = sns.barplot(data = df1, x='type', y='counts',hue='type',legend='full')
for i in aa.containers:
  plt.bar_label(i,)
plt.legend()
plt.title("Distribution of Type in dataset")
plt.show()
```

# Top 10 Movies Producing Countries

```python
# top 10 countries producing highest movies
top_10_countries = data.groupby('country').agg(counts=('show_id','count')).sort_values('counts',ascending = False).head(10)
top_10_countries

plt.figure(figsize=(12,6))
ax1 = sns.barplot(top_10_countries, x='country',y='counts', hue='country')
for i in ax1.containers:
  plt.bar_label(i,)
plt.title("Top 10 Highest Movies producing Countries")
plt.grid(alpha=0.20)
plt.xticks(rotation = 45)
plt.show()
```

# Shortest and Longest Movie durations

```python
movies = data[data['type']=='Movie'].reset_index()
movies.sample(3)
```

```python
# creating a feature for minutes for identifying the duration correctly

minutes = []
for i in range(len(movies)):
  iter = str(movies['duration'][i])
  if iter=='NaN' or iter=='nan':
    min = 0
    minutes.append(min)
  else:
    min = iter.split(' ')[0]
    minutes.append(int(min))

movies['minutes'] = minutes
movies.minutes
```

```python
max = movies.minutes.max()
min = movies[movies['minutes']!=0]['minutes'].min()
shortest_movie_name = movies[movies['minutes']==min]
print(f"Shortest duration movie {min} minutes -->> {shortest_movie_name.iat[0,3]}\n")
longest_movie = movies[movies['minutes']==max]
print(f"Longest duration movie {max} minutes -->> {longest_movie.iat[0,3]}")
```
